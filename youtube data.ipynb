```
# https://stackoverflow.com/questions/1563967/generate-sql-statements-with-python

def read(table, **kwargs):
    """ Generates SQL for a SELECT statement matching the kwargs passed. """
    sql = list()
    sql.append("SELECT * FROM %s " % table)
    if kwargs:
        sql.append("WHERE " + " AND ".join("%s = '%s'" % (k, v) for k, v in kwargs.iteritems()))
    sql.append(";")
    return "".join(sql)

def delete(table, **kwargs):
    """ deletes rows from table where **kwargs match """
    sql = list()
    sql.append("DELETE FROM %s " % table)
    sql.append("WHERE " + " AND ".join("%s = '%s'" % (k, v) for k, v in kwargs.iteritems()))
    sql.append(";")
    return "".join(sql)
```

```
def vids(kw="overwatch"):
  _, dat = youtube_search(kw)
  accounts = []
  accounts_visited = set()
  
  def to_doc(obj):
    return {
        "title": obj["snippet"]["title"],
        "url": obj["id"]["videoId"],
        "by": obj["snippet"]["channelId"]
    }
  def add_account(obj):
    aid = obj["snippet"]["channelId"]
    aname = obj["snippet"]["channelTitle"]
    if aid in accounts_visited:
      return
    else:
      accounts_visited.add(aid)
      accounts.append({
          "name": aname,
          "url": aid
      })
  list(map(add_account, dat))
  return list(map(to_doc, dat)), accounts

  

vids, accs = vids()
```

```
def upsert(table, **kwargs):
    """ update/insert rows into objects table (update if the row already exists)
        given the key-value pairs in kwargs """
    keys = ["%s" % k for k in kwargs]
    values = ["'%s'" % v for v in kwargs.values()]
    sql = list()
    sql.append("INSERT INTO %s (" % table)
    sql.append(", ".join(keys))
    sql.append(") VALUES (")
    sql.append(", ".join(values))
    sql.append(") ON DUPLICATE KEY UPDATE ")
    sql.append(", ".join("%s = '%s'" % (k, v) for k, v in kwargs.items()))
    sql.append(";")
    return "".join(sql)

  
def post_to_query(o):
  return ("INSERT INTO posts(title, url, posted_at, author) "
         "SELECT '" + o["title"] + "', '" + o["url"] + "', 0, id "
         "FROM accounts "
         "WHERE url = '" + o["by"] + "' "
         "limit 1 ;")
  
acc_inserts = list(map(lambda a: upsert("accounts", **a), accs))
vids_inserts = list(map(post_to_query, vids))
```
